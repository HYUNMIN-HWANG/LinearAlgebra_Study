```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(1,3,'o')   # x, y, circle
plt.plot(1,3,'s')   # x, y, square
plt.plot(1,3,'ro')   # x, y, red circle
plt.show()
```

```python
x = np.arange(-9,10)    # vector
print(x)

y = x**2    # vector
print(y)

plt.plot(x,y,'r')
plt.plot(x,y/2,'gs')    # gs : green square
plt.show()
```

```python
plt.plot([0,3],[-1,1], label='first line')  # [start, end], [start, end]
plt.plot([-2,0],[-4,1], label='second line')
plt.legend()    # show labels
plt.show()
```
