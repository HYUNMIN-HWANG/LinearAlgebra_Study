# Vector-scaler multiplication
- Vectors : An 'ordered' list of numbers 
- scaler : simgle number


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vec = [3,4,5,2] # vector -> vector가 우리가 생각하는 vector가 아니었던 것
s = 2   # scale 

print(vec*s)    # repeat vector 
```

```python
vec = np.array([3,4,5,2]) # vector
s = 3   # scale 

print(vec*s)    # repeat vector 
```

```python
vec2d = np.array([1,2])
s1 = 2
s2 = .5
s3 = -1

# indexing
vec2d[0]    # start counting zero
```

```python
plt.plot([0, vec2d[0]],[0,vec2d[1]],'bs-')  # created a bash line between start and end points
plt.plot([0, s1*vec2d[0]],[0, s1*vec2d[1]],'ro-', label='v*s1') 
plt.plot([0, s2*vec2d[0]],[0, s2*vec2d[1]],'kp-', label='v*s2') 
plt.plot([0, s3*vec2d[0]],[0, s3*vec2d[1]],'g*-', label='v*s3') 

plt.axis('square')  # 판을 정사각형으로 만들어준다.
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```

- assignment        
create two vectors (2D vectors)     
-> plot each vectors      
-> plot v1+v2, v1-v2, (v1*4+v2/2)    


```python
v1 = np.array(np.random.randint(0,10, size=(20)))
v2 = np.array(np.random.randint(-20,5, size=(20)))

print(v1)
print(v2)
# print(v1[0], v1[1])

plt.plot([0,v1[0]], [0,v1[-1]] ,'s-')
plt.plot([0,v2[0]], [0,v1[-1]] ,'ro-')
plt.show()
```

```python
sum_result = v1 + v2
sub_result = v1 - v2
new_result = v1*4+v2/2

plt.plot([0,sum_result[0]], [0,sum_result[-1]] ,'bo-',label='sum_result')
plt.plot([0,sub_result[0]], [0,sub_result[-1]] ,'ro-',label='sub_result')
plt.plot([0,new_result[0]], [0,new_result[-1]] ,'go-',label='new_result')
plt.legend()
plt.show()
```
