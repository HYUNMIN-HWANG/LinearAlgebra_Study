# Linear Transformation     
: n차원의 벡터를 m차원의 벡터로 transforamation 하는 것으로 생각하자.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)

xx, yy = np.meshgrid(x, y)
plt.scatter(xx, yy, s=20, c=xx+yy)
plt.show()
```

The Linear Transformation associated with a Matrix

```python
# transformation matrix T
T = np.array([[-1,0],[0,-1]])
print(T.shape)
T
```

```python
xy =  np.vstack([xx.flatten(), yy.flatten()])
xy.shape
```

```python
T @ xy[:, 0]
```

```python
trans = T @ xy
trans.shape
```

```python
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)
```

```python
f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

---

## Stretching and Rotation

```python
# rotation
T = np.array([
    [1.3, -2.4],
    [0.1, 2]
])

trans = T @ xy

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

```python
# 90도 rotation
T = np.array([
    [0, -1],
    [1, 0]
])


trans = T @ xy

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

```python
# Stretching
T = np.array([
    [5, 0],
    [0, 5]
])


trans = T @ xy

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

```python
# Contracting
T = np.array([
    [0.1, 0],
    [0, 0.1]
])


trans = T @ xy

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```
